<h1>displaying multiple plots </h1>

 important --- >> subplot(nrows , ncolumn ,plot no.)

```python
import matplotlib.pyplot as plt 
import numpy as np

#plot1 
x = np.array([0,1,2,3])
y =np.array([3,8,1,10])

plt.subplot(1,2,1)  # subplot() helps draw  multiple plots in one 
plt.plot(x,y)

#plot 2 :
x= np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)   # subplot(nrows , ncolumn ,plot no.)
plt.plot(x,y)

plt.show()
```

THE SUBPLOT () FUNCTION

the subplot function takes three arguments that describes the layout of the figure 

the layout is organised in rows and columns , which are represented by the first and second argument 


plt.subplot(1,2,1)

// the figure has 1 row , 2 columns , and this plot is the first plot .



plt.subplot(1,2,2)
// has 1 row , 2 columns , plot no. second

each color has a value that ranges from 0 to 100


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y , c = colors , cmap = 'viridis') #built in colour map

plt.show ()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x , y , c= colors , cmap= 'viridis')

plt.colorbar()   # this will add the color bar to the plot 

plt.show()
```

<h1> combine color size and alpha 

```python
# best visualized when the dots are transparent 

x = np.random.randint(100 , size =(100))  # 100 values
y = np.random.randint(100 ,size=(100))
colors = np.random.randint(100 , size = (100))
sizes = 10*np.random.randint(100 , size=(100))
plt.scatter(x , y , c = colors , s = sizes  , alpha = 0.5 , cmap = 'nipy_spectral')

plt.colorbar()

plt.show()
```

<h1> PIE CHART <h2>



```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

by default the plotting starts from the x axis called wedge and moves counterclockwise 


size of each wedge is determined by comaring the values with other values using : x / sum(x)

LABELS

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

my_labels = [ 'armas' , 'kriti' ,'khushi' , 'ava']

plt.pie( y , labels = mylabels)
plt.show()
```

<h4>start angle </h4>


changing the default start angle 

default is 0 degrees 

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ['apples', "Bananas", "Cherries", "Dates"]

plt.pie(y , labels  = mylabels  , startangle= 90)
plt.show()
```

EXPLODE

each value is represented how far from the centre each wedge is displayed :



```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0 , 0.5 , 0, 0]

plt.pie( y , labels = mylabels , explode= myexplode)
plt.show()
```

SHADOW 

Add a shadow to the pie chart by setting the shadow parameter to true 

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y , labels = mylabels  , explode = myexplode ,shadow= True)

plt.show()
```

COLORS --> you can set the color of each wedge using the colors parameter 


```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ['red' ,'yellow' , 'purple' , 'brown']

plt.pie(y , labels = mylabels , colors = mycolors )
plt.show()
```
