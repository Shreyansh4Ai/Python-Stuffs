Format String FMT


marker | line | colour

```python
ypoints = np.array([3,4,6,8,9])

# plt.plot(ypoints , 'o:r')
#plt.plot(ypoints, '*-.c')

plt.plot(ypoints, marker = '*' ,ms = 20) # marker size
plt.show()
```

CREATE LABELS FOR A PLOT 

```python
x = np.array([12,25,85,74,36,91,0])

y = np.array([52,14,70,91,0,4,52])

plt.plot(x, marker = 'o')
plt.plot(y ,marker = 'o' ,color ='g')

plt.xlabel("Average pulses ")
plt.ylabel("calories consumed ")


plt.show()
```

CREATE A TITLE FOR A PLOT 

```python
x = np.array([40,26,30,90])

y = np.array([50,80,40,20])
#z =np.array([10,20,30,40])
plt.plot(x,y ,color = 'b', marker = 'o')

plt.title("supra bang bang ")

plt.xlabel("torque per ms")
plt.ylabel("fuel consumed")
plt.show()
```

SET FONT PROPERTIES FOR TITLE AND LABELS 

```python
x = np.array([10,20,30,40,50])

y = np.array([60,50,40,30,20])

font1 = {'family':'serif' , 'color':'blue' , 'size':20}

font2 = {'family':'serif' , 'color':'green' ,'size':15}

plt.title('sports watches' , fontdict= font1)

plt.xlabel("average scores" ,fontdict = font2)

plt.ylabel("games" , fontdict =font2)

plt.plot(x,y)
plt.show()
```
