SET FONT PROPERTIES FOR TITLE AND LABELS 

```python
x = np.array([10,20,30,40,50])

y = np.array([60,50,40,30,20])

font1 = {'family':'serif' , 'color':'blue' , 'size':20}

font2 = {'family':'serif' , 'color':'green' ,'size':15}

plt.title('sports watches' , fontdict= font1)

plt.xlabel("average scores" ,fontdict = font2)

plt.ylabel("games" , fontdict =font2)

plt.plot(x,y)
plt.show()
```

<h1>displaying multiple plots </h1>

 important --- >> subplot(nrows , ncolumn ,plot no.)

```python
import matplotlib.pyplot as plt 
import numpy as np

#plot1 
x = np.array([0,1,2,3])
y =np.array([3,8,1,10])

plt.subplot(1,2,1)  # subplot() helps draw  multiple plots in one 
plt.plot(x,y)

#plot 2 :
x= np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)   # subplot(nrows , ncolumn ,plot no.)
plt.plot(x,y)

plt.show()
```

THE SUBPLOT () FUNCTION

the subplot function takes three arguments that describes the layout of the figure 

the layout is organised in rows and columns , which are represented by the first and second argument 


plt.subplot(1,2,1)

// the figure has 1 row , 2 columns , and this plot is the first plot .



plt.subplot(1,2,2)
// has 1 row , 2 columns , plot no. second
