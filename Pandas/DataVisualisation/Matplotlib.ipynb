<h1> Matplotlib </h1>

```python
import matplotlib.pyplot as plt 

import numpy as np


xpoints =np.array([1,5])
ypoints = np.array([3,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
x = np.array([1,3,6,4,8,0])
y = np.array([1,2,3,4,5,6])

plt.plot(x,y)
plt.show()
```

```python
#only x or y

x=np.array([3,6,9,3])

plt.plot(x , 'o')  # for no line only points 

plt.show()
```

MARKERS

```python
ypoints = np.array([4,6,8,9,0])

 # plt.plot(ypoints , 'o')
plt.plot(ypoints , marker = '*')
plt.show()
```

Format String FMT


marker | line | colour

```python
ypoints = np.array([3,4,6,8,9])

# plt.plot(ypoints , 'o:r')
#plt.plot(ypoints, '*-.c')

plt.plot(ypoints, marker = '*' ,ms = 20) # marker size
plt.show()
```

MARKER COLOUR 

```python
ypoint = np.array([2,5,6,8])

plt.plot(ypoint , marker ='o' , ms = 30 , markeredgecolor = 'r'  , markerfacecolor = 'y')
plt.show()
```

<h1> Matplotlib line </h1>

linestyle or shorter ls

```python
ypoints = np.array([3,9,4,5,8])

plt.plot(ypoints , linestyle = 'dotted')
plt.show()
```

```python
ypoints = np.array([3,9,4,5,8])

plt.plot(ypoints , linestyle = 'dashed')
plt.show()
```
